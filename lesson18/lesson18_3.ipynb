```python
import numpy as np
import pandas as pd
from pandas import Series

subjects = ['國文', '英文', '數學', '地理', '歷史']

scores = np.random.randint(50, 101, size=(50, 5))
scores_df = pd.DataFrame(scores, columns=subjects, index=range(1, 51))
scores_sum = scores_df.sum(axis=1)
scores_average = scores_df.mean(axis=1)

display(scores_df)
```

```python
from pandas import Series
#在表格底下列位新增項目
def to_statistic(value:Series) -> Series:
    return Series([value.max(),value.min(),value.mean()], index=['最高分','最低分','平均分'])
scores_df.apply(to_statistic)
```
