```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
scores_df = pd.DataFrame(scores,columns=['國文','英文','數學','地理','歷史'],index=range(1,51))
scores_df
```

```python
# 使用dataframe統計的方法,sum,mean,max,min,median,count,std
scores_sum = scores_df.sum(axis=1) #axis為0則直向欄計算，1為橫向列計算
#scores_average = scores_sum / 5.0
scores_average = scores_df.mean(axis=1)
scores_average
scores_df['總分'] = scores_sum
scores_df['平均'] = scores_average
scores_df
```

```python
#scores_df[['國文','英文','數學','地理','歷史']]
#自訂function
from pandas import Series
def get_level(s:Series):
    average = s.mean()
    if average < 60:
        return '不及格'
    elif average < 80:
        return '甲'
    else:
        return '優'

scores_df['等級'] = scores_df[['國文','英文','數學','地理','歷史']].apply(get_level,axis=1)
```
