```python
import matplotlib.pyplot as plt

student1 = {'chinese':78, 'english': 80, 'history':75, 'discover': 85}
student2 = {'chinese':69, 'english': 92, 'history':69, 'discover': 80}

subjects=list(student1.keys())
scores1=list(student1.values())
scores2=list(student2.values())

fig=plt.figure()
axes=fig.add_subplot(1,1,1)
axes.plot(subjects,scores1,'r^-.',label='stu1')
axes.plot(subjects,scores2,'go--',label='stu2') #可做簡化呼喚方式變換圖型標示
axes.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
student1 = {'chinese':78, 'english': 80, 'history':75, 'discover': 85}
student2 = {'chinese':69, 'english': 92, 'history':69, 'discover': 80}

subjects = list(student1.keys())
scores1 = list(student1.values())
scores2 = list(student2.values())

fig = plt.figure()
axes = fig.add_subplot(1,1,1)
axes.plot(subjects, scores1,color='red', marker='o',linewidth=3, markersize=13, label='stu1')
axes.plot(subjects, scores2,color='blue', marker='o',linewidth=3, markersize=13, label='stu2')
axes.legend()
plt.show()
```
