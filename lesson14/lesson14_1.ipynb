```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

#codespace
font = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
#windows系統指令
#font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
#============#
#mac系統指令
#font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)

student1 = {'國文':78, '英文': 80, '數學':75, '歷史': 85, '探索':92}
student2 = {'國文':90, '英文': 75, '數學':98, '歷史': 65, '探索':75}

subjects=list(student1.keys())
scores1=list(student1.values())
scores2=list(student2.values())

fig=plt.figure(figsize=(8, 5),dpi=250,facecolor='#cccccc') #後面color為兩個c一組進制編碼RGB三色
axes=fig.add_subplot(1,1,1)
axes.plot(subjects,scores1,'r^-.',label='學生1')
axes.plot(subjects,scores2,'go--',label='學生2')
axes.set_xticklabels(subjects,fontproperties=font) #將X軸中文字型顯示出來，不使用此行會只顯示空白文字無法顯示
axes.set_title("學生1和學生2分數比較表",fontproperties=font)
axes.set_xlabel("科目",fontproperties=font)
axes.set_ylabel("分數",fontproperties=font) #只要有中文字型都需要追加 fontproperties=font進行設定
axes.legend(prop=font) #因plot內的label是中文字型，所以需要追加指令設定 prop=font 顯示中文

#display(axes.get_xticks()) #顯示X軸的位置數值
xtick_nums = axes.get_xticks()
for index in range(len(xtick_nums)): #進行X軸對應的Y數值抓取顯示數值字串而非圖表
#    print('學生1')
#    print(xtick_nums[index],scores1[index]) #第一種方式指令抓取但只是用純字串顯示十組資料
#    print('學生2')
#    print(xtick_nums[index],scores2[index])  
    axes.annotate(str(scores1[index]),(xtick_nums[index]+0.1,scores1[index])) #直接在圖表上顯示該點位的數值
    axes.annotate(str(scores2[index]),(xtick_nums[index],scores2[index]+0.5)) # +後面則是在顯示的數值進行位移顯示
plt.show()
```
