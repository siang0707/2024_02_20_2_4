```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, num=50)

y1 = x * 2 + 1 #由圖表了解此為線性關係
y2 = x ** 2

figure, axes = plt.subplots()
axes.plot(x, y1)
axes.plot(x, y2, color='red', linewidth=1.0, linestyle='--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, num=50)

y1 = x * 2 + 1 #由圖表了解此為線性關係
y2 = x ** 2

figure, axes = plt.subplots()
axes.plot(x, y1)
axes.plot(x, y2, color='red', linewidth=1.0, linestyle='--')
axes.set_xlim((-1, 1)) #顯示局部X軸的-1到1之間的狀態
axes.set_ylim((-2, 2)) #顯示局部的Y軸-2到2之間的狀態
plt.show()
```

```python
import math
array_x = np.arange(0, 2.5, 0.1) #顯示0到2.5，以0.1為一個進制，但不包含2.5
display(array_x.shape)

y1 = np.sin(math.pi * array_x)
y2 = np.cos(math.pi * array_x)

figure, axes = plt.subplots()
axes.plot(array_x, y1, 'b*')
axes.plot(array_x, y2, 'g^')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = [('apple', 5000), ('banna', 7000), ('orange', 3000), ('tomato', 4000), ('guava', 6000)]
data_df = pd.DataFrame(data, columns=['fruit', 'numbers'])
display(data_df)

figure, axes = plt.subplots()
x_array = data_df['fruit'].values
y_array = data_df['numbers'].values

axes.bar(x_array, y_array)
x_ticks_value = axes.get_xticks() #抓取X軸的對應位置編號
for index in range(len(data)):
    print(x_ticks_value[index], y_array[index])
    axes.annotate(str(y_array[index]),(x_ticks_value[index]-0.2,y_array[index]+100))

plt.show()
```
