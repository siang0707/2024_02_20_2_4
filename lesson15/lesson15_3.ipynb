```python
#Dictionary資料型式
import pandas as pd

data = {
    '縣市': ['台北', '台北', '台北', '新竹', '新竹', '新竹'],
    '年份': [2000, 2001, 2002, 2001, 2002, 2003],
    '人口': [2.2, 2.5, 2.7, 1.1, 1.5, 2.0]
}

df1 = pd.DataFrame(data)
display(df1)
```

```python
#多個相同索引Series, 可以組成1個DataFrame
display(df1['縣市'])
display(df1['年份'])
display(df1['人口'])
```

```python
#二維的numpy array
import numpy as np

data = np.random.randint(50, 101, size=(50,5)) #50列5欄
scores_df = pd.DataFrame(data, columns=['國文', '英文', '數學', '地理', '歷史'], index=range(1, 51))
display(scores_df)
```

```python
total_series = scores_df['國文'] + scores_df['數學'] + scores_df['英文'] + scores_df['地理'] + scores_df['歷史']
display(total_series)
display(scores_df)
scores_df['總分'] = total_series #可直接追加到陣列當中新增出欄位
display(scores_df)
scores_df['平均'] = total_series / 5.0
display(scores_df)
```

```python
display(scores_df.loc[10]) 
display(scores_df.iloc[9])
```

```python
#slice
display(scores_df.loc[1:10])
#使用list
display(scores_df.loc[[1, 3, 5, 7, 9]])
#抓前10筆
display(scores_df.iloc[:10]) #索引型式
#抓最後10筆
display(scores_df.iloc[-10:])
```

```python
from requests import Response
import requests
from io import StringIO
import csv
import pandas as pd

def get_data()->list[dict]:
    link = 'https://raw.githubusercontent.com/roberthsu2003/__2024_02_20_2_4__/main/lesson11/students.csv'
    try :
        down:Response=requests.get(link)
        
    except:
        print("Erro")
        return 1
    file:str=down.text
    b:StringIO=StringIO(file)
    cs=csv.DictReader(b)
    data:list[dict]=list(cs)
    return data

data:list[dict] = get_data()
pd.DataFrame(data)
```
