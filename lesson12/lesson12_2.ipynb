```python
#原本csv檔改為讀取json檔
import requests
from requests import Response
def get_response() -> Response | None:
    json_url = 'https://data.moenv.gov.tw/api/v2/aqx_p_488?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=datacreationdate desc&format=JSON'
    try:
        response:Response = requests.get(json_url)
        response.raise_for_status()
        if response.status_code == 200: #有.ok 跟 .status_code 這兩種方式判定
            print("下載成功")
            return response
        else:
            print("下載失敗")
            return None
    except Exception as e: #Exception是檢查錯誤的涵義，然後回傳至設定的e當中
        print(e)
        print("連線失敗")
        return None
```

```python
response:Response | None = get_response()
if response:
    rawString:str = response.text #內為json格式，不是csv格式
    print(rawString)
```

```python
from pydantic import BaseModel,Field,field_validator

class Site(BaseModel):
    站點名稱:str = Field(alias='sitename')
    縣市:str = Field(alias='county')
    aqi:float
    時間:str = Field(alias='datacreationdate')
    pm25:float = Field(alias='pm2.5')

    @field_validator("pm25",mode='before')
    @classmethod
    def empty_to_zero(cls, value):
        if value == '':
            return '0.0'
        else:
            return value

class Sites(BaseModel): #不再是用RoorModel，而是BaseModel
    records:list[Site]
```

```python
data:Sites = Sites.model_validate_json(rawString) #此model為可以驗證json格式
data
#for item in data.records: #不是用data:Sites這個，而是使用上方的第二class BaseModel
#    print(item)
```

```python
download_data:list[dict] = data.model_dump()['records'] #傳換為python資料結構
download_data
```

```python
import pandas as pd
pd.DataFrame(download_data) #再藉由pandas將python資料結構轉會為表格形式
```
